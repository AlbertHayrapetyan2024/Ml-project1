```python
import pandas as pd
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("PS_20174392719_1491204439457_log.csv", nrows=10000)

# Preparing the Data
X = pd.get_dummies(df.drop(columns=['isFraud']))
y = df['isFraud']

# Model
model = DecisionTreeClassifier()

# Data sharing to test and train parts
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  # Corrected variable name

# Teach a model
model.fit(X_train, y_train)

# Predict a model
predictions = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)

#Data visualisation
figure = px.pie(df, names = 'type', title = "Data visualization")
figure.show()

print("Accuracy:", accuracy)
```
